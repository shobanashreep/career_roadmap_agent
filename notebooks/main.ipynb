```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()
```

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words("english"))
```

```python
def preprocess_text(text):

    words = text.split()

    words = [
        lemmatizer.lemmatize(word)
        for word in words
        if word not in stop_words
    ]

    return " ".join(words)
```

```python
df["processed_text"] = df["clean_text"].apply(preprocess_text)
```

```python
import re

def extract_skills(text):

    found_skills = []

    for skill in skills:

        pattern = r'\b' + re.escape(skill) + r'\b'

        if re.search(pattern, text):
            found_skills.append(skill)

    return found_skills
```

```python
df["combined_text"] = (
    df["Job Title"].str.lower()
    + " "
    + df["processed_text"]
)
```

```python
df["skills"] = df["combined_text"].apply(extract_skills)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer(
    max_features=5000
)
```

```python
tfidf_matrix = tfidf.fit_transform(
    df["processed_text"]
)
```

```python
print(tfidf_matrix.shape)
```

```python
len(tfidf.get_feature_names_out())
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
def recommend_jobs(user_input, top_n=5):

    user_vector = tfidf.transform([user_input.lower()])

    similarities = cosine_similarity(
        user_vector,
        tfidf_matrix
    ).flatten()

    df_temp = df.copy()

    df_temp["score"] = similarities

    df_temp = df_temp.sort_values(
        by="score",
        ascending=False
    )

    df_temp = df_temp.drop_duplicates(
        subset="Job Title"
    )

    return df_temp[
        ["Job Title", "skills", "score"]
    ].head(top_n)
```

```python
def find_skill_gap(user_skills, required_skills):

    user_skills = {
        skill.strip().lower()
        for skill in user_skills
    }

    required_skills = {
        skill.strip().lower()
        for skill in required_skills
    }

    missing = list(required_skills - user_skills)

    return sorted(missing)
```

```python
def generate_roadmap(missing_skills):

    beginner_skills = [
        "python",
        "sql",
        "statistics",
        "pandas",
        "numpy",
        "excel",
        "git"
    ]

    intermediate_skills = [
        "scikit learn",
        "machine learning",
        "django",
        "flask",
        "react",
        "angular",
        "nodejs",
        "flutter",
        "android",
        "ios",
        "firebase",
        "mongodb",
        "mysql",
        "postgresql"
    ]

    advanced_skills = [
        "deep learning",
        "tensorflow",
        "pytorch",
        "nlp",
        "aws",
        "azure",
        "docker",
        "kubernetes",
        "spark",
        "hadoop",
        "microservices"
    ]

    roadmap = {
        "Beginner": {
            "duration": "1-2 Months",
            "skills": []
        },
        "Intermediate": {
            "duration": "2-4 Months",
            "skills": []
        },
        "Advanced": {
            "duration": "4-6 Months",
            "skills": []
        }
    }

    for skill in missing_skills:

        if skill in beginner_skills:

            roadmap["Beginner"]["skills"].append(
                skill
            )

        elif skill in intermediate_skills:

            roadmap["Intermediate"]["skills"].append(
                skill
            )

        elif skill in advanced_skills:

            roadmap["Advanced"]["skills"].append(
                skill
            )

        else:

            roadmap["Intermediate"]["skills"].append(
                skill
            )

    return roadmap
```

```python
user_skills = [
    "python",
    "sql"
]

recommended_skills = [
    "python",
    "machine learning",
    "deep learning",
    "tensorflow",
    "pandas"
]

gap = find_skill_gap(
    user_skills,
    recommended_skills
)

print(gap)

print(
    generate_roadmap(gap)
)
```

```python
def career_roadmap(user_input, user_skills):

    recommendations = recommend_jobs(
        user_input,
        top_n=1
    )

    job_title = recommendations.iloc[0]["Job Title"]

    required_skills = recommendations.iloc[0]["skills"]

    missing_skills = find_skill_gap(
        user_skills,
        required_skills
    )

    roadmap = generate_roadmap(
        missing_skills
)

    courses = recommend_courses(
        missing_skills
    )

    return {
        "recommended_job": job_title,
        "required_skills": required_skills,
        "missing_skills": missing_skills,
        "roadmap": roadmap,
        "recommended_courses": courses
    }
```

```python
career_roadmap(
    user_input="python sql machine learning pandas",
    user_skills=["python", "sql"]
)
```

```python
course_map = {
    "python": "Python for Everybody",
    "sql": "SQL for Data Science",
    "pandas": "Data Analysis with Pandas",
    "numpy": "NumPy Fundamentals",
    "statistics": "Statistics for Data Science",
    "machine learning": "Machine Learning Specialization",
    "deep learning": "Deep Learning Specialization",
    "tensorflow": "TensorFlow Developer Certificate",
    "pytorch": "PyTorch for Deep Learning",
    "nlp": "Natural Language Processing Specialization",
    "django": "Django for Beginners",
    "react": "React Complete Guide",
    "flutter": "Flutter & Dart Complete Course",
    "aws": "AWS Cloud Practitioner",
    "docker": "Docker Essentials",
    "kubernetes": "Kubernetes Fundamentals",
    "scikit learn": "Scikit-Learn for Machine Learning"
}
```

```python
def recommend_courses(missing_skills):

    recommendations = {}

    for skill in missing_skills:

        if skill in course_map:
            recommendations[skill] = course_map[skill]

    return recommendations
```
